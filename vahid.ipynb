```python
university_towns=open(r"C:\Users\Desktop/university_towns.txt")
lines = university_towns.read().split("\n")

# After observinf the data we found out that the words in each state is not more then 16,
#so we are considering that in our text data if word lenth is less then 16 then it is a state
#and if more number of words are more then 16 then it is a region
    
state = []
states_regions = []
    
for element in lines:
    if len(element)<=16:
        # State
        state = element
    else:
        # Region
        states_regions.append({"State" : state, "Region" : element.split("(")[0].rstrip()})
            
```

```python
# Get GDP info: Drop unnessary header rows
gdp=pd.read_excel(r"C:\Downloads\gdplev.xlsx",header=6)
gdp
```

```python
# Take only quarterly info
gdp=gdp.iloc[:,4:7]
gdp
```

```python
gdp.drop(gdp.index[0],inplace=True)
gdp
```

```python
gdp.columns=["Quarter", "GDP current", "GDP chained"]
gdp.head()
```

```python
# Find GPD changes between quarters
gdp["Diff"]=gdp["GDP chained"].diff()
gdp.head()
```

```python
# Mark increase (1) or decrease (0) in GDP
gdp["Change"]=np.where(gdp["Diff"] < 0, "0", "1")
gdp
```

```python
# Now drop years before 2000 since "For this assignment, only look at GDP data from the first quarter of 2000 onward"
gdp = gdp[gdp.Quarter >= "2000q1"]
gdp.head()
```

```python
# recession year start when there is start 2 consicutive 0 in our column named change
# 2 columns refers to 2 consecutive quarters, if there in gdp fall in 2 quarters then it means recession start
change_str = ""
change_str = change_str.join(gdp["Change"].values)
change_str
```

```python
#finding the index for two 0 occuring together with two 11 following it
recession_index = change_str.index("0011")
```

```python
#here we are finding the start of recession and we are using +1 for finding 1 beofer pattern 0011
start_recession_index = change_str.rindex("1", 0, recession_index) + 1
recession_start= gdp.iloc[start_recession_index]["Quarter"]

recession_start
```

```python
# End recession is index of pattern + len of pattern
end_recession_index = recession_index + len("0011") - 1
recession_end= gdp.iloc[end_recession_index]["Quarter"]

recession_end
```

```python
#here we are finding the recession bottom ,that is quarter in recession with lowest gdp
recession_start=recession_start
recession_end=recession_end
    
df_recession = gdp[ (gdp.Quarter >= recession_start) & (gdp.Quarter <= recession_end) ]
c=df_recession[["GDP chained"]].idxmin()
    
recession_bottom=df_recession.loc[c,["Quarter"]]

recession_bottom
```
